```python
# Read XML
df_xml = (spark.read
          .format("xml")
          .option("rowTag", "item") # Adjust 'item' to the actual XML tag in your file
          .load("/Volumes/main/db_project/raw_data/chunks/chunk4_xml_temp"))

# Add Audit Columns and Save to Delta [cite: 51]
from pyspark.sql.functions import current_timestamp, lit

df_xml.withColumn("load_dt", current_timestamp()) \
      .withColumn("source", lit("XML_Source")) \
      .write.mode("append") \
      .saveAsTable("main.db_project.bronze_xml_data")
```
